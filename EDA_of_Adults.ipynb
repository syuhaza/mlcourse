### **Step 4: Explore Data Characteristics**

*  To explore whether an individual's income exceeds 50k a year involved selected countries
*  Based on the various attributes.
*  Target Variable: The income category (<=50k or >50k)

```python
#Examine total of sex feature
data['sex'].value_counts()
```

```python
# Average age of women
data[data["sex"] == "Female"]["age"].mean()
```

```python
# Proportion of German citizens (native-country feature)
float((data["native-country"] == "Germany").sum()) / data.shape[0]
```

```python
# Mean value and standard deviation of the age of those who receive more than 50K per year (salary feature) /
# and those who receive less than 50K per year

age1 = data[data["salary"] == ">50K"]["age"]
age2 = data[data["salary"] == "<=50K"]["age"]

print(f"The average age of those who receive more than 50K per year: Mean({round(age1.mean())}) Std({round(age1.std(), 1)})")
print(f"The average age of those who receive less than 50K per year: Mean({round(age2.mean())}) Std({round(age2.std(), 1)})")
```

```python
# It is true that people who earn more than 50K have at least high school education?
data[data["salary"] == ">50K"]["education"].unique()  # No
```

```python
# Display age statistic for each race and each gender
for (race, sex), sub_data in data.groupby(["race", "sex"]):
  print(f"Race: {race} | Sex: {sex}")
  #print("Race: {0}, Sex: {1}".format(race, sex))
  print(sub_data["age"].describe())
```

```python
# Proportion for status married men that gain more than 50K.

print("Married Men: ")
data[(data['sex'] == "Male") &
      (data["marital-status"].str.startswith("Married"))]["salary"].value_counts(normalize=True)
```

```python
# Proportion for status single men that gain more than 50K.

print("Single Men: ")
data[(data['sex'] == "Male") &
      ~(data["marital-status"].str.startswith("Married"))]["salary"].value_counts(normalize=True)
```

```python
# The maximum number of hours a person works per week
# Total people work such a number of hours
# The percentage of those who earn a lot among them

max_load = data["hours-per-week"].max()
print(f"Max time: {max_load} hours/week.")

num_workaholics = data[data["hours-per-week"] == max_load].shape[0]
print(f"Total number of such hard workers: {num_workaholics}.")

rich_share = (
    float(
        data[(data["hours-per-week"] == max_load) & (data["salary"] == ">50K")].shape[0]
    ) / num_workaholics
)
print(f"Percentage of rich among them: {int(100 * rich_share)}%")
```

```python
# Count the average time of work (hours-per-week) those who earning a little and a lot (salary) for each country (native-country)

for (country, salary), sub_data in data.groupby(["native-country", "salary"]):
  print(country, salary, round(sub_data["hours-per-week"].mean(), 2))
```

```python
# Another method
pd.crosstab(
    data["native-country"],
    data["salary"],
    values=data["hours-per-week"],
    aggfunc=np.mean
).T
```

```python
# The last data column, Churn is target variable
# The data type for Churn is bool, we will convert it into int with 'astype' method

data["Churn"] = data["Churn"].astype("int64")
```

### Sorting

```python
# Sorted by the values of one of the variables
data.sort_values(by="State", ascending=True).head()
```

```python
# Sort by multiple variables
data.sort_values(by=["Churn", "Total day charge"], ascending=[True, False]).head()
```

### Indexing & retrieving data

```python
# proportion of churned users
data["Churn"].mean()
```

```python
data.select_dtypes(include=np.number)[data['Churn'] == 1].mean()
```

```python
# Average churned user spend time on phone during daytime
data[data["Churn"] == 1]["Total day minutes"].mean()
```

```python
# Max length of international calls among loyal users for who don't have international plan
data[(data["Churn"] == 0) & (data["International plan"] == "No")]["Total intl minutes"].max()
```

```python
# indexing by name
data.loc[0:5, "State":"Area code"]
```

```python
# indexing by number
data.iloc[0:5, 0:4]
```

### Applying Functions

```python
data.apply(np.max)
```

```python
data[data["State"].apply(lambda state: state[0] == "W")].head()
```

```python
a = {"Yes":True , "No":False}

data["International plan"] = data["International plan"].map(a)
data.head()
```

```python
data = data.replace({"Voice mail plan": a})
data.head()
```

### Grouping

```python
# Exp: df.groupby(by=grouping_columns)[columns_to_show].function()
columns_to_show = ["Total day minutes", "Total eve minutes", "Total night minutes"]

data.groupby(["Churn"])[columns_to_show].describe(percentiles=[])
```

```python
data.groupby(["Churn"])[columns_to_show].agg(["mean", "std", "min", "max"])
```

### Summary tables

```python
pd.crosstab(data["Churn"], data["International plan"])
```

```python
pd.crosstab(data["Churn"], data["Voice mail plan"], normalize=True)
```

```python
data["Churn"].value_counts()
```

```python
data.pivot_table(
    ["Total day calls", "Total eve calls", "Total night calls"],
    ["Area code"],
    aggfunc="mean",
)
```
